```python
import pandas as pd
import joblib
from tensorflow.keras.models import load_model
from utils.data_utils import load_config
from utils.model_utils import predict_price_targets, generate_signal
config = load_config()
df = pd.read_csv('data/processed/features.csv', index_col='date', parse_dates=True)
lstm_model = load_model('models/lstm_shareindia.keras')
xgb_model = joblib.load('models/xgboost_shareindia.joblib')
forecast = predict_price_targets(lstm_model, xgb_model, df, config, None)
signal = generate_signal(forecast, df, config)
print('Signal Generated:', signal)
```
